```python
def get_group_idx(df):
  female_idx = df.index[df["sex_female"]==1]
  male_idx = df.index[df["sex_female"] == 0]
  graduate_idx = df.index[(df["university_education"]== df["highschool_education"])]
  university_idx = df.index[df["university_education"] == 1]
  highschool_idx = df.index[df["highschool_education"] == 1]
  other_education_idx = df.index[df["other_education"] == 1]
  married_idx = df.index[(df["single"]==df["marriage_other"])]
  single_idx = df.index[df["single"]==1]
  marriage_other_idx = df.index[df["marriage_other"]==1]
  under_fifty_idx = df.index[df["AGE"]<50]
  over_fifty_idx = df.index[df["AGE"] >= 50]

  return (female_idx, male_idx, graduate_idx, university_idx, highschool_idx, other_education_idx, married_idx, single_idx,marriage_other_idx,under_fifty_idx,over_fifty_idx)
```

```python
def confusion_matrix_metrics(y_test, predicted_labels, category):
  tn, fp, fn, tp = metrics.confusion_matrix(y_test, predicted_labels).ravel()
  fp_rate = fp/(fp+tn)
  fn_rate = fn/(fn+tp)
  frac_positive = (fp+tp)/(fn+fp+tn+tp)
  print("\tFalse Positive Rate for "+category+":", '%.4f'%(fp_rate))
  print("\tFalse Negative Rate for "+category+":", '%.4f'%(fn_rate))
  print("\tFraction of "+category+" classified positive:", '%.4f'%(frac_positive))
```

We attempted to use Shapley explainers to further break down the performance of our models, and gain insight into which features were most relevant to the performance of each of our four models. To do so, we utilized the SHAP framework for Python, and created plots for our models, demonstrating how each feature "added" to the model performance [7]. More specifically, we created violin plots for the random forest and logistic regression models.


Notably, we did **not include the SHAP analysis for random forests in our final project**. Our plots for the random forest model showed **no** insight beyond the plots that we produced from the feature importances attribute of the models, so these violin plots were not useful in providing more information.


For the Shapley analysis of logistic regression, which had the potential to be more interesting, we unfortunately also saw little meaningful to be derived from the dot plots. However, we conclude that:

2) The distribution of SHAP values (both with and without sensitive features) strongly resemble the distribution of the random forests' "feature_importance" plots. This is true EXCEPT for the "PAY_N" features, which matter very much in the random forest, but have near-zero SHAP values.

2A) We believe that this disparity in the importance of PAY_N features may be related to the multicollinearity of features, and the negative impacts this has on the performance of logistic regression. Notably, it is very hard (and potentially illogical) to figure out what exactly is wrong with the model, since we have violated one of logistic regression's key assumptions regarding multicollinearity.


We also think it is important to note, here, the importance of the *types* of plots that we show. For example, the bar graphs show only the MAGNITUDE of the SHAP values, without telling us their signs (which way they impact the outcome variable). The violin plots are helpful because they show which way they influence the outcome, but are much more difficult to understand. We decided to include both in this writeup, but would definitely choose to include *only* the bar plots in the future, because of their ease of understanding.

####  <span style="color: darkseagreen">**Limitations and Future Directions:**</span>

**Limitations of the data:** As we mentioned more thoroughly in our datasheet, our exploration of the data has made us really question the ethics, and overall quality, of the dataset we are using.


From an accuracy and precision standpoint, we realize that any of our work will be significantly limited by the quality of our data. We are particularly concerned about the ambiguity about wheteher or not the "ground truth" that we predict is really ground truth, or the output of another model.


From an ethics standpoint, our research question centers around fair algorithms, but this cannot happen without truly fair data (starting from collection). The lack of sourcing information, as well as information about data consent and collection, makes us question the ethics regarding this experiment.

Since consent cannot happen retroactively, the only thing we could do to improve our concerns with data quality are to attempt a similar analysis on a related dataset (ie: credit card defaulting) that was collected with subject consent.


**Limitations of the models:** Each of the modeling methods we explore make **significant assumptions about the data distribution,** which aren’t necessarily entirely valid. We will explore these for each model type below.

With our logistic regression models, the major limitation is that logistic regression learns linear boundaries. Based on our initial dataset exploration, we have no particular reason to believe that there is a linear boundary between those that default and those that do not. Based on the fairly good performance of the LR models, it seems that perhaps this was a reasonable assumption, but it is surely a major limitation.

LR models also assume independence of observations and a large sample size. Especially since our dataset was flawed, we are not sure that each observation was independent. Thinking about families and other potential social relationships, it is possible that some of these individuals impacted other observations. However, it is reasonable that *in general*, observations are independent, and we do know that we had quite a large sample size.

Finally, and perhaps most problematic, LR models assume that there is little to no  “multicollinearity,” of features. This means that for LR to work well, there should be little to no correlation of our features with the output. In reality, multiple independent variables are highly correlated with each other - spending amounts and paying patterns month to month are extremely related, and the inital allocation of credit is highly correlated with customers' past credit history.

One *future direction* that we could take here is exploring this multicollinearity more in depth. Specifically, we would really like to explore the initial allocation of credit (denoted as "LIMIT_BAL" in the dataset) and its relation to other features. The setting of this credit limit is itself an algorithmic (or at least human-made) decision, and thus it is irresponsible to view at as a set truth. In future work, we would like to see if perhaps we can predict the initial allocation of credit from demographic features.

Random forests have far fewer assumptions, namely because they are non-parametric models. Indeed, random forests (as bagged, ensemble methods) are robust to outliers and nonlinearity, and typically have a low risk of overfitting. However, they might overfit if there is particularly noisy data. We have no particular reason to think that our models were overfitting though. Also, if we did, we could limit overfitting by setting the MAX_DEPTH argument for the models at a lower value.



**Limitations of the analysis:** More could be done to explore the existing models quantitatively. For example, the main "sensitive feature" that we are evaluating currently is gender. To begin, we could also carry out similar confusion matrix analyses for the other sensitive features (age, education level, etc.). Indeed, this might even be more interesting because rather than a binary (as gender is modeled in this data), these features are multiclass (different levels of education, different age tiers).

Even more importantly, we could explore the performance of our current model more rigorously by looking at intersectional identities. For example, rather than looking at the performance for women and the performance for college-educated customers as independent groups, what is the performance for college-educated women? This is an essential part of truly "fair" or just algorithmic design.

**Other decisions/limitations:**

Simply the decisions associated with plotting have a huge impact on the quality of our report. Even with the exact same models and data, some plots were far easier to understand, and some plotting choices made model performance appear much better or worse than it really was. This was especially relevant to generating SHAP plots.

#### <span style="color: darkseagreen">**A Final Note:**</span>

Most importantly, we must not extrapolate the results of this study to larger social contexts. We must constrain our claims to within the context of this very small, specific dataset. We cannot claim that "credit card defaulting can be predicted with comparable accuracy with or without inclusion of sensitive features," but we can perhaps make this claim within the context of this one bank in Taiwan in 2005. This is a major flaw with much research, so we want to emphasize caution in making large, extrapolated claims.
